# 1. Importing Dataset

```python
import pandas as pd
data = pd.read_csv('Projec_final.csv')
data.info()
data
```

# 2. Setting up Environment

```python
from pycaret.classification import setup

# Configurar o ambiente PyCaret
exp = setup(data, target='Enviroment', session_id=123)

data.head()
```

# 3. Compare Models

```python
from pycaret.classification import compare_models
compare_models()
```

# 4. Create Model

```python
from pycaret.classification import create_model
knn = create_model('knn')
```

# 5. Tune Model

```python
from pycaret.classification import tune_model
tuned_knn = tune_model(knn)
```

```python
print(tuned_knn)
```

```python
print(tuned_knn)
```

# 6. Ensemble Model

```python
from pycaret.classification import ensemble_model
bagged_knn = ensemble_model(knn)
```

```python
# Change the method to 'Bagging'
bagged_knn = ensemble_model(knn, method = 'Bagging')
```

# 7. Blend Models

```python
knn = create_model('knn', verbose=False)
```

```python
!pip install pycaret
```

```python
from pycaret.classification import *
blended_model = blend_models(estimator_list=[knn], method='soft')
```
